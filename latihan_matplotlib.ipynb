```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.figure(figsize = (12,5))
plt.plot(x, y, 'r')

plt.xlabel('This is X', fontsize =15)
plt.xticks([1,2,3,4,5], ['Bart', 'Lisa', 'Jack', 'Andi', 'As'])

plt.ylabel('This is Y', fontsize = 10)
plt.yticks([1,2,3,4,5])

plt.title('This is title')

plt.show()
```

```python
plt.plot(x, y, 'r--')
plt.plot(y, x, 'g*')
```

```python
plt.figure(figsize = (12,8))

plt.subplot(2,2,1)
plt.plot(x, y, 'r--')
plt.title('Ini subplot 1')
plt.axis('off')

plt.subplot(2,2,2)
plt.plot(y, x, 'g*')
plt.title('Ini subplot 2')

plt.subplot(2,2,3)
plt.plot(x,y)
```

```python
fig = plt.figure(facecolor = 'yellow')

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'b')
axes1.set_xlabel('Ini X1')
axes1.set_ylabel('Ini Y1')
axes1.set_title('Ini Title 1')

axes2.plot(y, x, 'r')
axes2.set_xlabel('Ini X2')
axes2.set_ylabel('Ini Y2')
axes2.set_title('Ini Title 2')
axes2.set_axis_off()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols =2, figsize = (8,8), facecolor = 'yellow')
for axe in axes:
    for ax in axe: 
        ax.plot(x, y, 'b')
        ax.set_xlabel('Ini X')
        ax.set_ylabel('Ini Y')
        ax.set_title('Ini Title')
# axes
# axes[0][0].plot(x,y,'b', label = 'Ini 1')
# axes[0][0].plot(y,x,'--r', label = 'Ini 2')
# axes[0][0].legend()
plt.tight_layout()
```
