## Soal 2

```python
mpg[mpg['horsepower'].isna()]
```

```python
horse_na = mpg[mpg['horsepower'].isna()].index
```

```python
mpg.groupby('origin').mean()['horsepower']
```

```python
origin_mean = mpg.groupby('origin').mean()['horsepower']
def fillna_origin(cols):
    if pd.isna(cols['horsepower']):
        return round(origin_mean[cols['origin']])
    else:
        return cols['horsepower']
```

```python
mpg[['horsepower', 'origin']].apply(print, axis = 1)
```

```python
mpg['horsepower'] = mpg[['horsepower', 'origin']].apply(fillna_origin, axis =1)
```

```python
mpg.loc[horse_na]
```

## Soal 3

```python
mpg['origin'].unique()
```

```python
#If the horsepower is lower and equal to 105, then it is lower. If it is higher than 105 then it is upper
def add_horse_origin(x):
    if x['horsepower'] <=105:
        return 'lower {}'.format(x['origin'])
    else:
        return 'upper {}'.format(x['origin'])
```

```python
mpg['horse_origin'] = mpg[['horsepower', 'origin']].apply(add_horse_origin, axis = 1)
```

```python
mpg['horse_origin'].value_counts()
```
